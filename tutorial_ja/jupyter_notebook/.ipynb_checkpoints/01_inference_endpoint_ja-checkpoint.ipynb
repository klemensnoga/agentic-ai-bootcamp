## 学習目標

このモジュールを修了すると、以下のことができるようになります：
- NVIDIA Inference Microservices（NIMs）とは何か、および高速化された AI inference における役割を理解する
- NVIDIA API Catalog から NVIDIA API Key を取得して設定する
- Python の requests ライブラリを使って、NIM の cloud または local endpoint に inference リクエストを送信する
- OpenAI-compatible な NIM API からの chat completion レスポンスを解析・解釈する

## NVIDIA Inference Microservices（NIMs）の使い方

NIMs は、多様な microservice にオンラインでアクセスできるプラットフォームである [NVIDIA API Catalog](https://build.nvidia.com/explore/discover) で公開されている、使いやすい open API を通じて素早くアクセスできます。NIMs を使い始めるには、登録が必要な `NVIDIA API Key` が必要です。下のスクリーンショットに示すように `ログインボタンをクリックしてメールアドレスを入力する` ことで登録でき、その後の手順に従うか、[NVIDIA NGC](https://ngc.nvidia.com/signin) の登録（*アカウント名 -> setup -> Generate Personal Key をクリック*）を通じて API Key を生成することもできます。手続きが完了したら、今後使用できる場所に API Key を保存してください。API Key のサンプルは `nvapi-` から始まり、アンダースコア `_` を含む 64 文字が続きます。

すでにアカウントをお持ちの場合は、以下の手順で NVIDIA API KEY を取得してください：

- [こちら](https://build.nvidia.com/explore/discover) からアカウントにログインします。
- 希望するモデルをクリックします。
- Input で Python タブを選択し、Get API Key をクリックしてから Generate Key をクリックします。
- 生成されたキーを NVIDIA_API_KEY としてコピー・保存します。これで endpoint へのアクセスが可能になります。

<div style="text-align: center;">
  <img src="images/nim-catalog.png" style="width: 900px; height: auto;">
</div>


````python
import os
import getpass
import warnings
warnings.filterwarnings("ignore")
if not os.environ.get("NVIDIA_API_KEY", "").startswith("nvapi-"):
    nvapi_key = getpass.getpass("Enter your NVIDIA API key: ")
    assert nvapi_key.startswith("nvapi-"), f"{nvapi_key[:5]}... is not a valid key"
    os.environ["NVIDIA_API_KEY"] = nvapi_key
    os.environ["NGC_API_KEY"] = nvapi_key
````

## はじめに

このラボは、以下のいずれかの方法で実施できます：

- **Cloud Endpoint** – Cloud GPU 上でホストされた NVIDIA NIMs にアクセスする
- **Local Endpoint** – 自分の GPU 上に NIMs をローカルにデプロイする

## Cloud Endpoint

````python
import requests
import json
from pprint import pprint

url = "https://integrate.api.nvidia.com/v1/chat/completions"

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {os.environ['NVIDIA_API_KEY']}"
}

payload = {
    "model": "meta/llama-3.2-3b-instruct",
    "messages": [{"role": "system", "content": "What is 2+2?"}],
    "temperature": 0.6,
    "top_p": 0.95,
    "max_tokens": 4096,
    "frequency_penalty": 0,
    "presence_penalty": 0,
    "stream": False
}

response = requests.post(url, headers=headers, json=payload)
````

````python
pprint(response.json())
````

#### モデルアーティファクトのキャッシュのセットアップ

NIMs はハードウェア上で最大パフォーマンスを達成するために最適なプロファイルが選択されるよう、多数のファイルをダウンロードします。モデルアーティファクトをキャッシュする場所を `LOCAL_NIM_CACHE` として設定し、変数をエクスポートしてください。

````python
from os.path import expanduser
home = expanduser("~")
os.environ['LOCAL_NIM_CACHE']=f"{home}/.cache/nim"
!echo $LOCAL_NIM_CACHE
````

````python
!mkdir -p "$LOCAL_NIM_CACHE"
# !chmod 777 "$LOCAL_NIM_CACHE"
````

### NIM LLM Microservice の起動

以下のセルで docker run コマンドを実行して、NIM LLM microservice を起動します。`CONTAINER_PORT` と `NIM_CONTAINER_NAME` の実行時値は次のセルで割り当てられるため、各セッションが独自のホストポートとコンテナ名を使用できます。

```python
docker run -it --rm -d --gpus all --name=$NIM_CONTAINER_NAME --shm-size=16GB  -e NGC_API_KEY  -v '$LOCAL_NIM_CACHE':/opt/nim/.cache  -u $(id -u) -p $CONTAINER_PORT:8000 nvcr.io/nim/meta/llama-3.2-3b-instruct:latest
```

この Docker コマンドは、以下のフラグを使用して NIM LLM microservice を起動します：

- `-it`: 疑似 TTY を割り当て、インタラクティブなプロセス用に STDIN を開き続ける
- `--rm`: コンテナが終了したときに自動的に削除する
- `-d`: コンテナをデタッチモード（バックグラウンド）で実行する
- `--gpus all`: コンテナがすべての利用可能な GPU にアクセスできるようにする
- `--name=$NIM_CONTAINER_NAME`: 現在のユーザーと選択したホストポートを使ってコンテナに名前を付ける
- `--shm-size=16GB`: /dev/shm のサイズを 16GB に設定する
- `-e NGC_API_KEY`: NGC_API_KEY 環境変数をコンテナに渡す
- `-v $LOCAL_NIM_CACHE:/opt/nim/.cache`: ローカルの NIM キャッシュディレクトリをコンテナ内の /opt/nim/.cache にマウントする
- `-u $(id -u)`: 現在のユーザーの UID でコンテナを実行する
- `-p $CONTAINER_PORT:8000`: 動的に選択されたホストポートをコンテナのポート 8000 にマップする
- `nvcr.io/nim/meta/llama-3.2-3b-instruct:latest`: 使用する Docker イメージを指定する

システムには複数のプロセスが動作している可能性があるため、ノートブックは使用中のホストポートと共有コンテナ名の両方を再利用しないようにする必要があります。以下のコードは、空きポートを検索して、現在のユーザーセッション用にユニークなコンテナ名を生成します：

````python
import getpass
import os
import random
import re
import socket

def find_available_port(start=11000, end=11999):
    while True:
        # Randomly select a port between start and end range
        port = random.randint(start, end)
        
        # Try to create a socket and bind to the port
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
            try:
                sock.bind(("localhost", port))
                # If binding is successful, the port is free
                return port
            except OSError:
                # If binding fails, the port is in use, continue to the next iteration
                continue

def sanitize_container_component(value: str) -> str:
    sanitized = re.sub(r"[^a-zA-Z0-9_.-]+", "-", value.lower()).strip(".-_")
    return sanitized or "user"

username = sanitize_container_component(getpass.getuser())
os.environ['CONTAINER_PORT'] = str(find_available_port())
os.environ['NIM_CONTAINER_NAME'] = f"llm-nim-{username}-{os.environ['CONTAINER_PORT']}"

print(f"Your have been allotted the available port: {os.environ['CONTAINER_PORT']}")
print(f"Your NIM container name is: {os.environ['NIM_CONTAINER_NAME']}")
````

````python
! docker run -it -d --rm \
--gpus "device=${CUDA_VISIBLE_DEVICES}" \
--name=$NIM_CONTAINER_NAME \
--shm-size=16GB  \
-e NGC_API_KEY \
-v $LOCAL_NIM_CACHE:/opt/nim/.cache \
-u $(id -u) \
-p $CONTAINER_PORT:8000 \
nvcr.io/nim/meta/llama-3.2-3b-instruct:latest

# In order to ensure, the local NIM container is completely loaded and doesn't remain in pending stage, we instantiate a wait interval
! sleep 60
````

### クイックテストの実行
NIM が起動・稼働していることを、以下の 2 つの方法でクイックテストできます：
- LangChain NVIDIA Endpoints
- シンプルな OpenAI completion リクエスト

**パラメーター説明：**
- **base_url**: NIM docker イメージがデプロイされている URL。
- **model**: デプロイされた NIM モデルの名前。
- **temperature**: サンプリングのランダム性を調整するパラメーター。温度を下げると、高い確率を持つ単語が選ばれやすくなります。
- **top_p**: モデルの決定性を制御するパラメーター。正確で事実に基づく回答を求める場合は低く設定し、より多様な回答を求める場合は高い値に設定します。
- **max_tokens**: 生成される出力 token の最大数。

````python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

llm = ChatNVIDIA(base_url="http://0.0.0.0:{}/v1".format(os.environ['CONTAINER_PORT']), model="meta/llama-3.2-3b-instruct", temperature=0.1, max_tokens=1000, top_p=1.0)

result = llm.invoke("What is 2+2?")
print(result.content)
````

エラーが出力された場合は、しばらく待ってから上のセルを再実行してください。NIM コンテナが完全に起動していないことが原因の可能性があります。

````python
!curl -X 'POST' \
    "http://0.0.0.0:${CONTAINER_PORT}/v1/completions" \
    -H "accept: application/json" \
    -H "Content-Type: application/json" \
    -d '{"model": "meta/llama-3.2-3b-instruct", "prompt": "What is 2+2?", "max_tokens": 64, \
    "temperature": 0.1, "max_tokens": 1000, "top_p": 1.0, "stop": ["."]}'
````
