# Agentic AI Bootcamp
---

## 学習目標

1. NVIDIA Inference Microservices (NIMs) の機能と、高速化された AI inference のために NIM endpoint を使用する方法を理解する。
2. Model Context Protocol (MCP) と、それがツールおよびデータソースの標準化された通信をどのように実現するかを理解する。
3. tool 定義と input schema を用いたlow level の Server API を使って MCP サーバーを作成する。
4. NVIDIA NIM を LLM バックエンドとして LangGraph の StateGraph を使ったステートフルなチャットボットワークフローを構築し、パース可能なレスポンスのために structured output を実装する。
5. NVIDIA Nemo Agent Toolkit を使って MCP サーバーと NIM endpoint を接続し、intent-routed な agent ワークフローを構築する。
6. 学んだ概念をチャレンジ演習に応用して、完全な agentic アプリケーションを構築する。
7. チャレンジで完成させた Nemo Agent Toolkit ワークフローを活用する agent を OpenClaw 上で作成する。

## デジタルコンシェルジュ

#### プロローグ:質問が多すぎる店

Harmony Records は 20 年間音楽を販売してきました — レコード、CD、そして最終的にはデジタルダウンロード。会社がついにオンラインストアを立ち上げたとき、メールが殺到し始めました。顧客はどのアルバムが在庫にあるか、なぜ請求書の内容がおかしく見えるのか、後悔した買い物を返金できるか、といった質問を投げかけてきました。サポートチームは圧倒されていました。

ストアマネージャーの Elena は、AI agent について耳にしていました。FAQ の回答を繰り返すだけの単純なチャットボットではなく、データを参照し、アクションを実行し、本当に不確実な場合には人間に引き継ぐ、本物の推論システムです。彼女はそれが欲しいと思いました。Marcus というソリューションアーキテクトを呼び寄せました。

「彼女を Aria と呼ぶことにしよう」と Elena は言いました。「サポートチームが一日の半分を費やしている業務をすべて彼女に任せる」。

Marcus は微笑みました。「彼女はチャットボット以上の存在になる。正しく作り上げよう」。

---

#### 第一章:声を見つける

Aria に最初に必要だったのは声でした — 顧客が文字通りに入力した内容ではなく、本当に意図していることを理解できる、十分に鋭い language model です。

Marcus は NVIDIA の inference endpoint への接続から始めました。コンテナ化され、事前に最適化された language model に、標準的な API 経由でアクセスできるものです。彼はレスポンスをテストし、モデルの呼び出し方法をチューニングしました。token がどう流れ込み、回答がどう流れ出るかを学び、クラウドで動作するモデルと、ストアが完全に制御するローカルハードウェアにデプロイされるモデルの違いを学びました。

Aria の声が形になっていきました。彼女は質問を読み、首尾一貫した返答を生成できるようになりました。しかし記憶はなく、ストアのデータへのアクセスもなく、行動する能力もありませんでした。雄弁ではあるが、役に立ちませんでした。

「手のない声だな」と Marcus は言いました。「彼女に道具を与えよう」。

---

#### 第二章:道具の使い方を学ぶ

ストアのデータベースにはすべての情報がありました:顧客、請求書、アルバム、アーティスト、トラック、購入履歴。しかしモデルは直接これをクエリできませんでした。Marcus は、これらの機能を Aria に対して個別の呼び出し可能なアクション — 必要に応じて Aria が呼び出せるもの — として公開する方法が必要でした。

彼は **Model Context Protocol** を発見しました。まさにこの目的のために設計されたオープン標準です。MCP は契約のように機能します:server はどの tool が利用可能か、それぞれが入力として何を期待するかを定義します。client として動作する Aria の推論層は、それらの tool を発見し、会話中に呼び出すことができます。

Marcus は high-level デコレーターを使って最初の MCP server を素早く構築しました。ほとんど追加の手間なく、普通の Python 関数を tool に変えるものです。彼はシンプルな chat ループでテストしました。Aria は今、アーティストを名前で検索できるようになりました。トラックリストを取得できました。特定のレーベルが何枚のアルバムをリリースしたかを数えられました。

しかし Aria が実際の顧客リクエストを — 多数を同時に、信頼できるネットワーク上で — 処理することを想像したとき、Marcus はもっと制御が必要だと知りました。彼は invoice 処理 server を low-level で再構築しました:明示的な input schema、慎重な validation、意図的なエラーハンドリング、そしてネットワーク上のどこからでも到達できる HTTP transport 層です。high-level アプローチはプロトタイピングには優雅でした。low-level アプローチは本番環境が要求するものでした。

Aria は今、手を得ました。しかしまだ計画の立て方を知りませんでした。

---

#### 第三章:ステップで考えることを学ぶ

ある顧客が書きました:*「先週アルバムを間違ったバージョンで購入してしまって、返金してほしいのですが、キャンセルする前にオリジナルプレスの年を教えてもらえますか?」*

この単一のメッセージには少なくとも 3 つのものが含まれていました:製品履歴に関する質問、条件、そして返金リクエスト。単一の tool 呼び出しでは処理できません。Aria はステップで推論する必要がありました — 情報を取得し、メモリに保持し、次に何をすべきかを決定し、タスクが完了するまでループする。

Marcus は **LangGraph** を導入しました。ステートフルな推論グラフを構築するためのフレームワークです。彼は Aria の思考を有向グラフとしてモデル化しました:node はアクション(LLM の呼び出し、tool の呼び出し、専門家へのルーティング)で、edge はそれらをつなぐロジックです。state — Aria が現在の会話について知っていることすべて — はこのグラフを流れ、各ステップで更新されます。

今、Aria はマルチターンの会話を扱えるようになりました。メッセージのスレッドが context を蓄積します。tool を呼び出して部分的な情報が返ってきた場合、彼女は止まることなく推論を続けられます。会話が一時停止した場合 — 人間のスーパーバイザーが大きな返金を承認する必要があるなど — 彼女は待ち、中断したところから正確に再開できます。

Marcus は重要な機能を追加しました:**agent skill** です。Aria が起動時に音楽データベースの完全な知識をメモリにロードするのではなく、skill 説明のディレクトリを与えられます。彼女は必要なときに必要なものだけを読みます。顧客がトラックのメタデータについて尋ねたとき、彼女は関連する skill 指示を取得し、それを使って正しいデータベースクエリを組み立て、きれいに回答しました。

Aria は今、計画し、行動し、記憶し、一時停止できるようになりました。しかし Marcus は大量のグルーコードを書いていました — モデル呼び出しを tool 呼び出しに配線する Python ファイル、想定し得るすべてのパスに対して手書きされたルーティングロジック。脆くて拡張が難しいものでした。

---

#### 第四章:性格を宣言する

Marcus は Elena と将来について話しました。すでに機能している返金に関する質問やカタログ検索だけでなく — Aria が将来どうなるかについて。さらに多くの agent、さらに多くの tool、プロモーション推薦用の新しい agent、配送追跡用のもう 1 つの agent。

「追加するたびにコードの半分を書き直さなければならない」と Marcus は言いました。「もっと良い方法がある」。

彼は **NeMo Agent Toolkit** を発見しました。手作業で配線された Python ではなく、YAML ファイルで agent の構成全体を記述できるフレームワークです。モデル、tool、ルーティングロジック、接続すべき MCP server — すべてが宣言されます。Marcus が Aria の動作を調整したいとき、コードベースではなく設定ファイルを編集します。

彼は intent classifier を組み込みました — 各受信メッセージを読み、適切な agent にルーティングする前に、それがどのタイプのリクエストかを決定する、小さく高速なモデルです。一般的な音楽の質問は一方の道へ。invoice と返金リクエストは別の道へ。これらのカテゴリに当てはまらないものは、人間に引き継がれるようにフラグが立てられました。

彼はまた **Phoenix** を接続しました。Aria が下したすべての決定を記録する observability プラットフォームです:どの tool を呼び出したか、各ステップで LLM が何を言ったか、各段階にどれくらいの時間がかかったか。何かがうまくいかなかったとき、Marcus はトレースを再生して、推論がどこで崩れたかを正確に確認できます。

Aria は今、観測可能で、設定可能で、拡張可能になりました。

---

#### 第五章:初めての本番稼働日

Aria が稼働した朝、サポートチームは緊張して見守りました。最初の顧客メッセージは午前 9:04 に届きました。

*「こんにちは、カタログにある Miles Davis のすべてのアルバムを探しています」*

Intent classifier がメッセージを読みました:一般的なカタログの質問。QNA agent にルーティング。Aria は彼女のカタログ検索 skill を呼び出し、データベースをクエリし、12 枚のアルバムを取得し、2 秒以内にきれいなリストで返答しました。

午前 9:11:*「先月の請求書があるのですが、注文した商品が届いていません。返金してほしいです」*

Intent classifier:返金リクエスト。Refund agent にルーティング。Aria は invoice 検索 tool を呼び出し、注文を特定し、それが発送されていないことを確認し、返金ワークフローを開始しました。彼女は最終確定する前に、Marcus の設計通りに、人間のスーパーバイザーに承認のためサマリーを提示しました。

午前 9:47:*「現代のジャズの状況についてどう思いますか?」*

Intent classifier:スコープ外。Aria はカタログと注文に関する質問を専門としているため、他のことで助けが必要な場合は人間につなぐことができますと丁寧に応答しました。

正午までに、サポートチームが対応したエスカレーションは 12 件未満で、通常の 50 件から大幅に減っていました。

Elena が Marcus のデスクに立ち寄りました。「彼女、やってるね」。

「彼女は簡単な部分をこなしているだけだ」と Marcus は言いました。「面白い仕事はこれからだ — 質問がもっと難しくなり、エッジケースが積み重なり、顧客が我々が予想していなかったことを聞き始めたとき」。

彼は Phoenix のトレースビューを表示しました。Aria の決定グラフが画面上で光っていました — 各メッセージが分類、ルーティング、tool 呼び出し、応答を通過するたびに node が光ります。

「でも今、彼女が何をしているかを正確に見ることができる」と彼は言いました。「ということは、彼女に教えることもできる」。

---

#### エピローグ:Aria が彼らに教えたこと

Aria を作ることで、Marcus と Elena はアイデアの完全な道筋を通り抜けました。

彼らは、language model がどれほど有能であっても、tool なしでは不完全であることを学びました — そして、それらの tool をどのように公開するかが非常に重要であることも。デコレーターは速いです。Low-level schema は信頼できます。本番システムには結局両方が必要になります。

彼らは、agent はスクリプトではないことを学びました。context を保持し、決定をルーティングし、人間の入力のために一時停止し、きれいに再開する必要のある、ステートフルな推論プロセスです。目で見て trace できるグラフは、コードに埋もれたループよりも価値があります。

彼らは、手書きのロジックはスケールしないことを学びました。宣言的な設定 — agent が何をするかを記述し、すべてのステップをプログラミングするのではない — が、プロトタイプと保守可能なシステムの違いを生みました。

そして、observability はオプションではないことを学びました。トレースできない agent は改善できない agent です。

Aria は完璧ではありませんでした。彼女は時々珍しいリクエストを誤分類し、時々人間が介入する必要があり、Marcus がまだ少しいらいらする invoice schema のエッジケースについて意見を持っていました。しかし、彼女は本物で、稼働しており、改善されつつありました。

サポートチームは最初の金曜日、月曜日まで残る未回答のチケットの山なしで帰宅しました。
