# Challenge Overview
---

Modern e-commerce support is a strong use case for agentic systems. Customers ask broad product questions, request help with past purchases, and expect fast, accurate routing when they need refunds or other account support. This challenge brings those needs together into one practical workflow so students can apply the course concepts in a realistic customer support setting.  

In this course challenge, you will extend the provided music store support system into an intelligent e-commerce agent that understands customer needs and routes each request to the right workflow. The agent should answer general customer inquiries, process refund requests, and identify customer intent so it can deliver the appropriate support experience.  

From this challenge, participants should be able to
1) create MCP servers utilising the low level python sdk
2) package MCP servers into a format reusable by any MCP client
3) implement agent skills based on progressive disclosure
4) craft workflows with human-in-the-loop through langgraph
5) utilise Nvidia NIMs as the LLM backbone in agentic workflows

The workflow has 3 main parts, 
1) the intent classifier, 
2) the question answering (qna) agent for handling general customer inquiries and
3) the refund agent for processing refund requests.

To complete the challenge, you will
1) allow human intervention if intent is unknown
2) incorporate the use of MCP protocol in the refund agent so it can access invoice data
3) utilize agent skills in the qna agent for grounded customer support responses

### Assignment 4 - NeMo Agent Toolkit

*Ensure assignments 1, 2 and 3 are complete before starting on this assignment*

This assignment is scored out of ***5 points***.

In this assignment, you will create a `workflow.yaml` configuration for the [NeMo Agent Toolkit (NAT)](https://docs.nvidia.com/nemo/agent-toolkit/latest/index.html) that connects to the invoice MCP server and answers queries about invoices in the Chinook database.

Your workflow must define:
1. An **MCP client** function group that connects to the invoice MCP server
2. An **LLM** backed by an NVIDIA NIM endpoint
3. A **ReAct agent** workflow that uses the MCP tools and LLM to answer invoice-related questions

The evaluation runs your `workflow.yaml` against 5 test questions. Each question is scored 0 or 1.

The following folders/files for the NeMo Agent Toolkit challenge have been provided to you.  

```bash
nemo_agent_toolkit
├── README.md
└── workflow.yaml
```

The task is to complete the workflow configuration in the following file.
- [nemo_agent_toolkit/workflow.yaml](../../challenge/nemo_agent_toolkit/workflow.yaml)

#### Testing your solution

You can test your workflow locally by starting the invoice MCP server and running:

```bash
nat run --config_file workflow.yaml --input "<question>"
```
