# Bonus Challenge — OpenClaw Integration

This bonus challenge extends the main workshop challenge by integrating your completed solution — the MCP invoice server and NemoClaw Agent Toolkit workflow — into **OpenClaw**, a self-hosted AI assistant that can take action on your behalf across platforms.

Choose the setup path that matches your environment:

## [07a — Docker Setup](07_bonus_challenge_docker.ipynb)

Run OpenClaw inside a Docker container for a fully isolated environment. The challenge repo is mounted into the container so your completed solution files are available to the agent. Recommended when you want to avoid modifying the host system or need a reproducible, disposable setup.

**Key steps:** clone OpenClaw repo → mount challenge repo → pull Docker image → configure music-store agent via OpenClaw CLI and direct filesystem edits → start MCP and NAT servers inside the container → query agent via TUI → inspect session traces.

## [07b — Host Installation](07_bonus_challenge_host.ipynb)

Run OpenClaw directly on your host machine. This path installs OpenClaw natively via a shell script, configures it as a systemd service, and wires up the music-store agent skill to your NAT workflow endpoint. Best suited for Linux systems where you have full OS access.

**Key steps:** install OpenClaw CLI → configure music-store agent → start MCP and NAT servers on host → query agent via TUI or dashboard → inspect session traces.
